3] Explain the following terms

 The difference in np.array, np.asarray and np.asanyarray

 The difference between Deep copy and shallow copy

* Difference between np.array(), np.asarray(), and np.asanyarray()

ANS.

np.array():

This function creates a new array from a list or tuple, and it always copies the input data unless copy=False is passed. It can take options such as specifying data type or forcing a deep copy.

```python
arr1 = np.array([1, 2, 3])
```

np.asarray():

 This function converts the input to an array, but unlike np.array(), it doesn't copy the data if the input is already an array. It's more efficient when you don't want to create a new array.

```python
arr2 = np.asarray([1, 2, 3])
```

np.asanyarray():

 Similar to np.asarray(), but it will preserve the input type if it's a subclass of ndarray (e.g., matrix). It’s useful when you want to work with objects like matrices and don't want to cast them back to a basic array.

```python
arr3 = np.asanyarray([1, 2, 3])
```

* Difference between Deep Copy and Shallow Copy

ANS.

* Deep Copy:

A deep copy creates a new object and recursively copies all objects found in the original, creating independent copies. In NumPy, if you modify the original array, it will not affect the deep copy.

```python
arr = np.array([1, 2, 3])
arr_deep_copy = np.copy(arr)
```

* Shallow Copy:

A shallow copy creates a new object, but references to the original object’s elements are maintained. If you change elements in the original object, those changes will also appear in the shallow copy.

```python
arr = np.array([1, 2, 3])
arr_shallow_copy = arr.view()
```

 b) Label the x-axis as 'Date'.

 c) Set the title of the plot as 'Temperature and Humidity Over Time

 ANS.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Step 2: Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=35, size=len(date_range))  # Temperature between 15 and 35
humidity = np.random.uniform(low=30, high=90, size=len(date_range))  # Humidity between 30% and 90%

# Step 3: Create the DataFrame
data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
}
df = pd.DataFrame(data)

# Step 4: Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature', linewidth=2)
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity', linewidth=2)
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Step 5: Set the title
plt.title('Temperature and Humidity Over Time')
plt.grid()  # Optional: Add grid for better visibility
plt.show()
```

Random humidity values between 30% and 90% are generated similarly.

* Creating the DataFrame: A dictionary is created with the generated data, and then a Pandas DataFrame is constructed.

A figure and an axis (ax1) are created using plt.subplots().

The 'Temperature' data is plotted on the left y-axis with a red color.

15]  Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

 a) Plot a histogram of the data with 30 bins.

 b) Overlay a line plot representing the normal distribution's probability density function (PDF).

 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

 d) Set the title of the plot as 'Histogram with PDF Overlay'

 ANS.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Standard deviation = 1

# Step 2: Plot the histogram with 30 bins
plt.figure(figsize=(12, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay the PDF
# Calculate the PDF values
mean = np.mean(data)
std_dev = np.std(data)
pdf = norm.pdf(bins, mean, std_dev)

# Plot the PDF
plt.plot(bins, pdf, 'k', linewidth=2, label='PDF')

# Step 4: Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.grid()
plt.show()
```

Explanation:

* Generating Samples:

The np.random.normal function is used to generate 1000 samples from a normal distribution with a mean (loc) of 0 and a standard deviation (scale) of 1.

The np.random.seed(0) ensures that the random numbers generated are reproducible.

* Plotting the Histogram:

A histogram of the data is plotted using plt.hist, specifying 30 bins and setting density=True to normalize the histogram. The alpha parameter controls the transparency of the histogram bars.

* Overlaying the PDF:

The mean and standard deviation of the generated data are calculated using np.mean and np.std.

The PDF is computed using norm.pdf, which evaluates the normal distribution at each bin center.

The PDF is then overlaid on the histogram using plt.plot.

* Labeling and Title:

The x-axis and y-axis are labeled as 'Value' and 'Frequency/Probability', respectively.

The title of the plot is set to 'Histogram with PDF Overlay'.

A legend and grid are added for clarity.

17]  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot .

ANS.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)  # 100 random points from a normal distribution
y = np.random.randn(100)

# Step 2: Create a DataFrame to hold the data
data = pd.DataFrame({'X': x, 'Y': y})

# Step 3: Determine the quadrant for each point
def determine_quadrant(row):
    if row['X'] > 0 and row['Y'] > 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] > 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    elif row['X'] > 0 and row['Y'] < 0:
        return 'Quadrant 4'
    else:
        return 'Origin'

data['Quadrant'] = data.apply(determine_quadrant, axis=1)

# Step 4: Create a scatter plot with Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant', s=100)

# Step 5: Label the axes and set the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Step 6: Display legend and show the plot
plt.legend(title='Quadrants')
plt.grid()
plt.axhline(0, color='gray', lw=0.8, ls='--')  # Adding horizontal line at y=0
plt.axvline(0, color='gray', lw=0.8, ls='--')  # Adding vertical line at x=0
plt.show()
```

Explanation:

* Generating Random Arrays:

Random arrays x and y are generated using np.random.randn, which creates points drawn from a standard normal distribution.

* Creating a DataFrame:

A Pandas DataFrame is created to hold the random points along with their respective coordinates.

* Determining Quadrants:

A function determine_quadrant is defined to determine the quadrant for each point based on its x and y coordinates. This function is applied to each row of the DataFrame.


* Creating the Scatter Plot:

A scatter plot is generated using Seaborn's scatterplot function, where points are colored based on their assigned quadrant. The hue parameter is used to color the points, and style can also be applied if desired. The size of the points is set using the s parameter.

* Labeling Axes and Setting the Title:

The x-axis and y-axis are labeled appropriately, and the plot title is set to 'Quadrant-wise Scatter Plot'.

* Legend and Grid:

A legend is added to show which colors correspond to which quadrants. Horizontal and vertical dashed lines are added at y=0 and x=0 to highlight the quadrants' divisions.

19]  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

ANS.



```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Prepare the output in the notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame for better manipulation
data = pd.DataFrame({'Category': categories, 'Value': values})

# Assign colors based on values and add to the DataFrame
data['Color'] = ['blue' if value < 50 else 'orange' for value in values]

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a figure for the plot
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400)

# Add hover tooltips to display exact values
hover_tool = HoverTool(tooltips=[("Category", "@Category"), ("Value", "@Value")])
p.add_tools(hover_tool)

# Create a bar chart
p.vbar(x='Category', top='Value', source=source, width=0.9,
       line_color='white', fill_color='Color')  # Use the 'Color' column from the source

# Show the plot
show(p)
```

21]  Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'

ANS.



```python
import plotly.graph_objs as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
values = np.random.randint(1, 100, size=5)  # Random values for the pie chart
labels = [f'Category {i+1}' for i in range(len(values))]  # Labels for the categories

# Create a pie chart
fig = go.Figure(data=[go.Pie(
    labels=labels,
    values=values,
    textinfo='percent+label',  # Show percentage and label
    hoverinfo='label+percent',  # Show label and percentage on hover
)])

# Update layout
fig.update_layout(title='Interactive Pie Chart')

# Show the plot
fig.show()
```

Explanation:

* Import Libraries:

Import the necessary libraries, including plotly.graph_objs for plotting and numpy for generating random data.

* Generate Data:

values:

An array of 5 random integers between 1 and 100, representing the values for each category in the pie chart.

labels:

 A list of labels for each category in the pie chart.

* Create the Pie Chart:

A Figure object is created, and a pie chart is added using go.Pie.

The textinfo parameter is set to display both percentages and labels on the pie slices.

The hoverinfo parameter specifies that both labels and percentages should be shown when hovering over the slices.

* Update Layout:

 The layout of the plot is updated to include the title.

* Display the Plot:

The show() function is called to display the interactive pie chart.
