import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

# Set random seed for reproducibility

np.random.seed(42)

# Generate synthetic dataset with correlated features

n = 100  # number of samples

x1 = np.random.normal(loc=0, scale=1, size=n)
x2 = x1 * 0.5 + np.random.normal(loc=0, scale=0.5, size=n)  # x2 is correlated
with x1

x3 = x1 * -0.3 + np.random.normal(loc=0, scale=0.5, size=n)  # x3 is negatively correlated with x1

x4 = np.random.normal(loc=1, scale=1, size=n)  # independent feature

# Create a DataFrame

df = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})

# Compute the correlation matrix

correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix

plt.figure(figsize=(10, 6))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={"shrink": .8})

plt.title('Correlation Matrix Heatmap')

plt.show()

import numpy as np

import plotly.express as px

# Generate sample data

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Aggregate sales data to get total sales per month and day

sales_heatmap = df.groupby(['Month', 'Day']).sum().reset_index()

# Create a pivot table for the heatmap

heatmap_data = sales_heatmap.pivot(index='Day', columns='Month', values='Sales')

# Create the heatmap using Plotly

fig = px.imshow(heatmap_data,
                labels=dict(x="Month", y="Day", color="Total Sales"),
                x=heatmap_data.columns,
                y=heatmap_data.index,
                color_continuous_scale='Viridis')

fig.update_layout(title='Sales Heatmap: Variation Across Months and Days',
                  xaxis_title='Month',
                  yaxis_title='Day')

fig.show()

import numpy as np

from bokeh.plotting import figure, show

from bokeh.io import output_notebook

from bokeh.transform import linear_cmap

from bokeh.colors import RGB

# Set up the output to display in a Jupyter Notebook

output_notebook()

# Generate random data

data_heatmap = np.random.rand(10, 10)

# Create x and y coordinates

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid

xx, yy = np.meshgrid(x, y)

# Prepare the data for the heatmap

data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Create a new figure

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=600, height=600)

# Define a color mapper

color_mapper = linear_cmap('value', palette='Viridis256', low=min(data['value']), high=max(data['value']))

# Create a heatmap using quadrilaterals

p.rect(x='x', y='y', width=0.1, height=0.1, source=data,
        fill_color=color_mapper, line_color=None)

# Show the plot

show(p)
